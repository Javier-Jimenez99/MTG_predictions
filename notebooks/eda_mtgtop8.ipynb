```python
from MTGpred.utils.database import get_all_decks
import pandas as pd

decks = get_all_decks(with_archetype=True)
df = pd.DataFrame(decks).drop(columns=["_id", "main_deck", "sideboard"])
print("There are {} decks in the database".format(len(df)))
```

## FORMAT

```python
df['format'].value_counts().plot(kind='barh')
```

## ARCHETYPE

```python
df['archetype'].value_counts().plot(kind='barh')
```

## DETAILED ARCHETYPE

```python
print(f"There are {len(df['detailed_archetype'].unique())} different detailed_archetypes")
df['detailed_archetype'].value_counts()[:50].plot.barh(figsize=(10, 10))
```

## NAME

```python
print(f"There are {len(df['name'].unique())} different names")
df['name'].value_counts()[:50].plot.barh(figsize=(10, 10))
```
